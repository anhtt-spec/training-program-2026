## 1. The Three Models: Waterfall vs Agile vs Hybrid

### Waterfall
- Fixed scope, rigid documentation, sequential phases
- Best for: compliance-heavy, well-defined projects

### Agile/Scrum
- Sprint-based, continuous feedback, adaptive
- Best for: evolving products, fast time-to-market

### Hybrid (Company Standard)
```
[Phase 1 — Waterfall]  Initiation & System Analysis
          ↓
[Phase 2 — Agile]      Iterative Development (Sprints)
          ↓
[Phase 3 — Waterfall]  System Test & Release
```

## 2. The P-I-N Technique + Buffer Management

| Step | Action | Tool |
|------|--------|------|
| **Plan** | Analyze requirements, create WBS | Claude + Jira |
| **Implement** | Execute tasks, track progress | Cursor/CodeX |
| **Next** | Review, commit, update status | Git + standup |

**Buffer Rule:** Add 20% buffer to any estimate. Never commit to 100% capacity.

````python
# Buffer Management Demo
tasks = [
    ('API endpoint /auth/login', 3),
    ('UI login form', 2),
    ('Input validation', 1),
    ('Unit tests', 1.5),
]
raw_total = sum(h for _, h in tasks)
buffered = raw_total * 1.2
print(f'Raw estimate: {raw_total}h')
print(f'With 20% buffer: {buffered}h')
for name, hrs in tasks:
    print(f'  • {name}: {hrs}h → buffered: {hrs*1.2:.1f}h')
````
